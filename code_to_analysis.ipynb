In this first step of analysis, we handling the missing value first for NaN value from logical type of features. We check the missing values for each features as below:

### Handling Duplicates

```python
# remove duplicates in dog_desc
dog_desc = dog_desc.drop_duplicates()
dog_travel = dog_travel.drop_duplicates()
```

### Encoding and Normalization

```python
# Encoding Binary Features

dog_travel = dog_travel.replace({True: 1, False: 0})
dog_desc = dog_desc.replace({True: 1, False: 0})
```
